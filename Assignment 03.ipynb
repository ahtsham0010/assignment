```python
def err_ranges(y, confidence=0.95):
    a = 1.0 * np.array(y)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
from statistics import NormalDist

def confidence_interval(y, confidence=0.95):
    dist = NormalDist.from_samples(y)
    z = NormalDist().inv_cdf((1 + confidence) / 2.)
    h = dist.stdev * z / ((len(y) - 1) ** .5)
    return dist.mean - h, dist.mean + h
```
